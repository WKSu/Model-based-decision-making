### Considering their end states, which outcomes of interests are negatively correlated?

Pairwise scatter plots of the end states quickly display the relation between the outcomes of interest. *pairs_scatter()* function of the EMA workbench can be used for this purpose.

Looking at the plot below,
- *Infection fraction* is zero for almost all scenarios in three policies.
- *Deceased population* and *maximum infection fraction* are positively correlated.
- *time of maximum* infection rate is negatively correlated with both *deceased population* and *maximum infection fraction*, meaning that the later the peak of the infection rate, the lower its magnitude, and the lower the *deceased popoulation*.  

```python
from ema_workbench.analysis.pairs_plotting import pairs_scatter
```

```python
fig, axes = pairs_scatter(results)
fig.set_figheight(8)
fig.set_figwidth(12)
plt.show()
```

### Considering only the 'no policy' scenarios, how do *infected fraction* and *deceased population* relate over time?   

*pairs_lines()* function of the workbench plots the pairwise relation between two timeseries outcomes over time.

Namely, in the plots below, each line depicts such a relation curve where 'o' is the starting point (time=0) and '+' is the end point. The results show that both *infected fraction* and *deceased population* increase in short term, yet *infected fraction* decreases later although *deceased population* keeps increasing. 

To plot the scenarios corresponding only to the 'no policy' case, we need to create a new tuple of results. We can do this by filtering the cases of interest both from experiments and outcomes based on their indices.

```python
from ema_workbench.analysis.pairs_plotting import pairs_lines

logical_index = experiments['policy'] == 'no policy' #logical_index is a boolean array.
new_experiments = experiments[logical_index]
new_outcomes = {key:outcomes[key][logical_index] for key in outcomes.keys()}
new_results = new_experiments, new_outcomes

fig, axes = pairs_lines(new_results)
fig.set_figheight(8)
fig.set_figwidth(12)
plt.show()
```

### Visualize the uncertainty range of *infected fraction* over time, together with individual scenarios

We can visualize the uncertainty range of an outcome indicator over time, i.e. the range between the minimum and maximum values at each time point, by using the *envelopes* function of the workbench. For this exercise, since we also would like to show some individual scenarios, we can use the *plot_lines_with_envelopes* function. We specify which scenarios we would like to show individually with the *experiments_to_show* argument.   

```python
from ema_workbench.analysis.plotting import plot_lines_with_envelopes

#iterate over the experiments and check one by one if both conditions are staisfied:
logical_index = np.array([outcomes['time of max'][i] < 10 and outcomes['max infection fraction'][i] > 0.4 for i in range(len(experiments))])

'''
#Alternatively,
#- Keep the truth values for each condition in a multidimensional list,
#- swap the axes and keep the truth values of all conditions for each experiment,
#- for each experiment, check if all conditions are True.

indices = [outcomes['time of max'] < 10, outcomes['max infection fraction'] > 0.4]
indices = np.swapaxes(indices, 0, 1)
logical_index2 = np.array([index.all() for index in indices])

'''

#Specifying 'outcomes_to_show' in the plot_lines_with_envelopes function rehapes the 'results' tuple. 
#To avoid this, we can use a copy of the results.  
new_results = experiments.copy(), outcomes.copy() 

fig, axes = plot_lines_with_envelopes(new_results,
                                      outcomes_to_show=['infected fraction R1'],
                                      experiments_to_show=logical_index)
plt.show()
```

### Visualize the outcome indicators maximum infection fraction, time of max, and the end state of deceased population on a single plot

Three outcome indicators can be displayed on a single plot by 3D scattering. 

Below is an example that differnetiates between policies by color, and the cases of interest by marker. In particular, the cases of interest are marked with a triangle, and the resat is marked with a circle.


```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'b', 'y']
markers = ['o', '^']
oois = ['max infection fraction', 'deceased population region 1','time of max']

for p, policy in enumerate(policies):
    newOutcomes = {key:value[experiments['policy']==policy] for key, value in outcomes.items()}
    newExperiments = experiments[experiments['policy']==policy]
    
    logical_index = [newOutcomes['infected fraction R1'][i, 99]>0.1 and newExperiments['infection ratio region 1'][i]>0.1 for i in range(len(newExperiments))]
    logical_index = np.array(logical_index)
    
    x_ = newOutcomes[oois[0]][logical_index]
    x = newOutcomes[oois[0]][~logical_index]
    ax.set_xlabel(oois[0])
    
    y_ = newOutcomes[oois[1]][logical_index,-1]
    y = newOutcomes[oois[1]][~logical_index,-1]
    ax.set_ylabel(oois[1])
    
    z_ = newOutcomes[oois[2]][logical_index]
    z = newOutcomes[oois[2]][~logical_index]
    ax.set_zlabel(oois[2])
    
    ax.scatter(x, y, z, c=colors[p], marker=markers[0], alpha=0.5, s=15, label=policy)
    ax.scatter(x_, y_, z_, c=colors[p], marker=markers[1], s=60, label=policy+'***')

ax.legend(loc=2, scatterpoints = 1)
ax.set_xlim([0, 0.5])
ax.set_ylim([0, 5e+07])
ax.set_zlim([0, 50])
fig.set_figheight(8)
fig.set_figwidth(12)
plt.show()
```
